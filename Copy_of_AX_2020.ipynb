```
from torchvision import datasets,models,transforms
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from PIL import Image

import time

import numpy as np
import matplotlib.pyplot as plt
import os

from google.colab import drive
drive.mount('/content/drive')
```

```
# Data augmentation part
# create a dict to save different kinds of data
transformer = {
    'train': transforms.Compose([
        transforms.RandomHorizontalFlip(),
        # transforms.RandomVerticalFlip()
        transforms.RandomResizedCrop(size=256, scale=(0.9, 1.0),ratio=(0.75, 1.3333333333333333)), # randomly resize pic
        transforms.RandomRotation(degrees=30),
        transforms.CenterCrop(size=224),
        transforms.ToTensor(),
        # Pytorch recommended: All pre-trained models expect input images normalized in the same way
        transforms.Normalize([0.485, 0.456, 0.406],# mean value for a 3-channel pic
                             [0.229, 0.224, 0.225])# std value for a 3-chennel pic
    ]),
    'valid': transforms.Compose([
        transforms.Resize(size=256),
        transforms.CenterCrop(size=224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ]),
    'test': transforms.Compose([
        transforms.Resize(size=256),
        transforms.CenterCrop(size=224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ])
}

```

```
# Load data
dataset_path = '/content/drive/My Drive/university_buildings'
train_directory = os.path.join(dataset_path, 'train')
valid_directory = os.path.join(dataset_path, 'valid')
test_directory = os.path.join(dataset_path, 'test')

all_data = {'train': datasets.ImageFolder(root = train_directory, transform = transformer['train']),
            'valid': datasets.ImageFolder(root = valid_directory, transform = transformer['valid']),
            'test': datasets.ImageFolder(root = test_directory, transform = transformer['test'])
            }

class_index = {}
classes_num = len(os.listdir(train_directory))-1
for k, v in all_data['train'].class_to_idx.items():
  class_index.update({v:k})
  
print(class_index)
# Create iterators for the Data loaded using DataLoader module


train_data_loader = DataLoader(all_data['train'], batch_size = 32, shuffle=True)
valid_data_loader = DataLoader(all_data['valid'], batch_size = 32, shuffle=True)
test_data_loader = DataLoader(all_data['test'], batch_size = 32, shuffle=True)
```

```
torch.cuda.get_device_name(0)
```

```
# Load and construct model
resnet50 = models.resnet50(pretrained = True)
resnet50 = resnet50.cuda()# .to('cuda:0')

# Freeze model parameters
# When a model is loaded in PyTorch, all its parameters have their ‘requires_grad‘ field set to true by default.
for k in resnet50.parameters():
    k.requires_grad = False

# Change the final layer of ResNet50 Model to full connected layer for Transfer Learning
# define custom network
resnet50.fc = nn.Sequential(
    nn.Linear(resnet50.fc.in_features, 256), # Layer 1
    nn.Sigmoid(), # Layer 2
    # nn.ReLU(),
    nn.Dropout(0.4), # Layer 3
    nn.Linear(256, len(os.listdir(train_directory))-1), # Layer 4 (num of classes of training set)
    nn.LogSoftmax(dim=1) # Layer 5 for using NLLLoss()
)

# Define Optimizer and Loss Function
loss_function = nn.NLLLoss(reduction='mean') # always used for muti-class classification task, and "sum" sum all loss in one batch
optimizer = optim.Adam(resnet50.parameters())

train_data_size = len(all_data['train'])
valid_data_size = len(all_data['valid'])
test_data_size = len(all_data['test'])

resnet50 = resnet50.cuda()
```

```
def train_and_validate(model, loss_f, optimizer, epochs=20):  
    model_log = []

    for epoch in range(epochs):
        print("Training... Epoch ",epoch+1,"out of ",epochs)
        
        # Set to training mode
        model.train()
        
        # Loss and Accuracy within the epoch
        train_loss = 0.0
        train_acc = 0.0
        
        valid_loss = 0.0
        valid_acc = 0.0
        
        for i, (inputs, labels) in enumerate(train_data_loader):
            inputs = inputs.to(device)
            labels = labels.to(device)
            optimizer.zero_grad()
            
            # Forward pass
            outputs = model(inputs)
            # Backpropagate the gradients
            loss = loss_f(outputs, labels)
            loss.backward()
            optimizer.step()
            
            # Calculate the loss
            train_loss += loss.item() * inputs.size(0)
            # Calculate the accuracy
            no, predictions = torch.max(outputs.data, 1)
            acc = 0
            for i in range(len(predictions)):
              if(predictions[i] == labels.data.view_as(predictions)[i]):
                acc += 1 
            train_acc += acc
            
        # Validation
        with torch.no_grad():

            # Set to evaluation mode
            model.eval()
            for n, (inputs, labels) in enumerate(valid_data_loader):
                inputs = inputs.to(device)
                labels = labels.to(device)

                # Forward pass
                outputs = model(inputs)
                # Calculate loss
                loss = loss_f(outputs, labels)
                valid_loss += loss.item() * inputs.size(0) # .item() retutrns the sum loss value of this batch;
                # Calculate acc
                no, predictions = torch.max(outputs.data, 1)
                acc = 0
                for i in range(len(predictions)):
                  if(predictions[i] == labels.data.view_as(predictions)[i]):
                    acc += 1 
                valid_acc += acc

        model_log.append([train_loss/train_data_size, valid_loss/valid_data_size, train_acc/train_data_size, valid_acc/valid_data_size])
        print("Epoch %d: \nTraining: Loss: %.4f; Accuracy: %.4f \nValidation: Loss: %.4f; Accuracy: %.4f" \
              % (epoch+1, train_loss/train_data_size, train_acc/train_data_size, valid_loss/valid_data_size, valid_acc/valid_data_size))
        torch.save(model, dataset_path + '/models/' + '_model_' + str(epoch) + '.pt')
            
    return model, model_log
```

```
def test(model, loss_f):

    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

    test_acc = 0.0
    test_loss = 0.0
    # Validation
    with torch.no_grad():
        model.eval()
        # Validation loop
        for j, (inputs, labels) in enumerate(test_data_loader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # Forward pass
            outputs = model(inputs)

            # Calculate loss
            loss = loss_f(outputs, labels)
            test_loss += loss.item() * inputs.size(0)

            # Calculate accuracy
            no, predictions = torch.max(outputs.data, 1)
            acc = 0
            for i in range(len(predictions)):
              if(predictions[i] == labels.data.view_as(predictions)[i]):
                acc += 1 
            test_acc += acc

            print("Test Batch number: %d Loss: %.4f, Accuracy: %.4f" % (j, loss.item(), acc / len(predictions)))

    avg_test_acc = test_acc / test_data_size

    print("Final test average accuracy : " + str(avg_test_acc))
```

```

def predict(model, test_image_name):
    
    transform = transformer['test']

    test_image = Image.open(test_image_name)
    plt.imshow(test_image)
    
    test_image_tensor = transform(test_image)

    if torch.cuda.is_available():
        test_image_tensor = test_image_tensor.view(1, 3, 224, 224).cuda()
    else:
        test_image_tensor = test_image_tensor.view(1, 3, 224, 224)
    
    with torch.no_grad():
        model.eval()
        # Model outputs log probabilities
        out = model(test_image_tensor)
        ps = torch.exp(out)
        topk, topclass = ps.topk(3, dim=1)
        for i in range(3):
            print("Predcition", i+1, ":", class_index[topclass.cpu().numpy()[0][i]], ", Score: ", topk.cpu().numpy()[0][i])
```
